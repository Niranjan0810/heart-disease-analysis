```python
# Heart Disease Data Analysis
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
## Load Dataset
```

```python
df = pd.read_csv("heart.csv")
```

```python
## Data Cleaning
```

```python
df = df.drop_duplicates()
```

```python
## Correlation Heatmap
```

```python
plt.figure(figsize=(12,8))
sns.heatmap(df.corr(), annot=True, fmt=".2f", cmap='coolwarm')

plt.title("Correlation Heatmap")
plt.xticks(rotation=45, ha='right')
plt.yticks(rotation=0)

plt.tight_layout()
plt.savefig("images/heatmap.png", dpi=300)
plt.show()
```

```python
## Age Distribution
```

```python
plt.figure(figsize=(8,5))
sns.histplot(df['age'], kde=True)

plt.title("Age Distribution")

plt.savefig("images/age_distribution.png", dpi=300)
plt.show()
```

```python
## Cholesterol vs Heart Disease
```

```python
plt.figure(figsize=(8,5))
sns.boxplot(x='target', y='chol', data=df)

plt.title("Cholesterol vs Heart Disease")

plt.savefig("images/cholesterol_vs_target.png", dpi=300)
plt.show()
```

```python
## Chest Pain Type vs Heart Disease
```

```python
plt.figure(figsize=(8,5))
sns.countplot(x=df['cp'], hue=df['target'])

plt.title("Chest Pain Type vs Heart Disease")

plt.savefig("images/cp_vs_target.png", dpi=300)
plt.show()
```

```python
## Max Heart Rate vs Heart Disease
```

```python
plt.figure(figsize=(8,5))
sns.boxplot(x='target', y='thalach', data=df)

plt.title("Max Heart Rate vs Heart Disease")

plt.savefig("images/thalach_vs_target.png", dpi=300)
plt.show()
```

```python
## Exercise-Induced Angina vs Heart Disease
```

```python
plt.figure(figsize=(8,5))
sns.countplot(x=df['exang'], hue=df['target'])

plt.title("Exercise-Induced Angina vs Heart Disease")

plt.savefig("images/exang_vs_target.png", dpi=300)
plt.show()
```
